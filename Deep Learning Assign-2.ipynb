```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy.linalg as LA
```

```python
def loss_function(x1,x2,l1,l2,e):
    loss = 0
    n = len(x1)
    for i in range(n):
        loss+=((l1-2)*x1[i]+(l2-3)*x2[i]-e[i])**2
    return loss
def plot_loss(x,y,z):
    ax = plt.axes(projection='3d')
    ax.plot_surface(x,y,z,color='b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()
```

```python
def main():
    x1 = np.random.uniform(-4,4,400)
    x2 = np.random.uniform(-4,4,400)
    e =np.random.normal(0,0.01,400)
    p1 = np.linspace(-30,30,50)
    p2 = np.linspace(-30,30,50)
    X,Y = np.meshgrid(p1,p2)
    Z = loss_function(x1,x2,X,Y,e)
    plot_loss(X,Y,Z)
    data = np.array((x1,x2,np.ones(400)))
    alpha = calculate_constant_lr(data)
    print(alpha)
    alphas = [.9*alpha/2,1.5*alpha/2,alpha,1.5*alpha]
    for alpha in alphas:
        print("Current learning rate",alpha)
        w = np.array((np.random.normal(),np.random.normal(),0))
        w0 = w
        w_list = []
        error_list = []
        ct = 0
        while True:
            loss = calculate_loss(w,data,e)
            gd = calculate_gradient(w,data,e)
            temp = w-alpha*gd
            if ct%100 == 0:
                print("Iteration Number:",ct,"  Current Error",loss)
                w_list.append(np.array((temp[0],temp[1])))
                error_list.append(loss)

            if loss < 0.001 :
                print("Gradient Descent Converged in Iterations:",ct)
                w_list.append(np.array((temp[0],temp[1])))
                error_list.append(loss)
                break

            w = temp
            ct+=1
        plt.plot(error_list)
        plt.title("For learning rate "+str(alpha))
        plt.show()
        w_list = np.array(w_list)
        z = loss_function(x1,x2,X,Y,e)
        plot(X,Y,z,w_list)
main()
```

```python
#part3
def calculate_constant_lr(data):
    Hess = 2*(np.dot(data,data.T))
    eigvals,_ = LA.eig(Hess)
    val = max(eigvals)
    val = 1/val
    return val
```

```python
#part4
def calculate_gradient(w,x,e):
    e = np.mean(e,axis=0)
    w = w-np.array((2,3,e))
    gd = 2*np.dot(x,np.dot(x.T,w))
    return gd
def calculate_loss(w,x,e):
    e = np.mean(e,axis=0)
    w = w-np.array((2,3,e))
    loss = np.dot(np.dot(w.T,x),np.dot(x.T,w))
    return loss
```

```python
def Adam():
    x1 = np.linspace(-3,3,250)
    x2 = np.linspace(-3,3,250)

    X,Y = np.meshgrid(x1,x2)
#     print(X.shape,Y.shape)
    X = X.astype('float128')
    Y = Y.astype('float128')
    w = [np.random.normal(),np.random.normal()]
    w0 = w
    alpha = 8*1e-3
    w = np.array(w,dtype='float128')
    w_list = []
    error_list = []

    ct = 0
    alpha = 8*1e-3
    b1 = 0.9
    b2 = 0.999
    e = 1e-8
    rep = [0]*2
    arr = [0]*2
    rep = np.array(rep,dtype='float128')
    arr = np.array(arr,dtype='float128')
    while True:
        loss = loss_function(w[0],w[1])
        temp = np.array((0,0),dtype='float128')
        update = np.array((0,0),dtype='float128')
        gd = calculate_gradient(w)
        gd = np.array(gd,dtype='float128')
        temp = b1*rep+(1-b1)*gd
        update = b2*arr+(1-b2)*np.square(gd)
        temp2 = temp/(1-b1**(ct+1))
        update2 = update/(1-b2**(ct+1))
        r  = np.multiply(alpha/(np.sqrt(update2)+e),temp2)
#         print("here",r)
        change = w-r
        if ct%100 == 0:
#             print("Iteration Number:",ct,"  Current Error",loss)
            w_list.append(change)
            error_list.append(loss)

        if loss < 0.001 or ct > 1000:
            print("Gradient Descent Converged in iterations:",ct)
            w_list.append(change)
            error_list.append(loss)
            break

        w = change
        rep = temp
        arr = update
        ct+=1

    print(loss)
    w_list = np.array(w_list)
    z = loss_function(X,Y)
#     print(z)
    plot(X,Y,z,w_list)
Adam()
```

```python
##Q-4
def loss_function(W,X,Y):
    loss = (np.dot(X,W)-Y)
    loss = np.dot(loss,loss)
    return loss
def calculate_gradient(W,X,Y):
    gd = np.dot(X.T,X)
    gd = np.dot(gd,W)
    gd -= np.dot(X.T,Y)
    gd = 2*gd
#     gd = 2*(np.dot(np.dot(X.T,X),W)-np.dot(X.T,Y))
    return gd
def main():
    mu1 = np.array([-3,4])
    mu2 = np.array([4,-3])
    cov_mat = np.array([[16,0],[0,9]])
#     print(mu1.shape)
    class_1 =np.random.multivariate_normal(mu1,cov_mat,300)
    class_2 = np.random.multivariate_normal(mu2,cov_mat,300)
    labels_1 = np.ones(200)
    labels_2 = np.full(200,-1)
    test_1 = class_1[200:]
    test_2 = class_2[200:]
    ytest_1=np.ones(100)
    ytest_2 = np.full(100,-1)
    class_1 = class_1[:200]
    class_2 = class_2[:200]
    train = np.hstack((np.vstack((class_1.T,np.ones(200))),np.vstack((class_2.T,np.ones(200)))))
    X_train = train.T
    Y_train = np.hstack((labels_1,labels_2))
    X_test = np.hstack((np.vstack((test_1.T,np.ones(100))),np.vstack((test_2.T,np.ones(100))))).T
    Y_test = np.hstack((ytest_1,ytest_2))
#     print(labels_1.shape)
    w = np.array((np.random.normal(),np.random.normal(),1))
    alpha = 5e-5
    error_list = []
    test_list = []
    w_list = []
    Max_iters = 100000
    for i in range(Max_iters):
        loss = loss_function(w,X_train,Y_train)
        gd = calculate_gradient(w,X_train,Y_train)
        w_update = w-alpha*gd
        error_list.append(loss)
        w_list.append(w)
        test_preds = np.dot(X_test,w)
        test_list.append(200-np.count_nonzero(np.sign(test_preds) == Y_test))
        if loss < 1e-3:
            break
        w = w_update
    print(loss)
    print("Gradient Descent After %d iterations"% i)
    print(w)
    plt.plot(error_list)
    plt.show()
    plt.plot(test_list)
    plt.show()
        
main()
```

```python
def normalise(x1,x2):
    u = np.mean(np.hstack((x1.T,x2.T)).T,axis = 0)
    sd = np.std(np.hstack((x1.T,x2.T)).T,axis = 0)
    for i in range(len(x1)):
        x1[i] = (x1[i]-u)/(sd)
        x2[i] = (x2[i]-u)/(sd) 
    return x1,x2
def main():
    mu1 = np.array([-3,4])
    mu2 = np.array([4,-3])
    cov_mat = np.array([[16,0],[0,9]])
#     print(mu1.shape)
    class_1 =np.random.multivariate_normal(mu1,cov_mat,300)
    class_2 = np.random.multivariate_normal(mu2,cov_mat,300)
    labels_1 = np.ones(200)
    labels_2 = np.full(200,-1)
    test_1 = class_1[200:]
    test_2 = class_2[200:]
    ytest_1=np.ones(100)
    ytest_2 = np.full(100,-1)
    class_1 = class_1[:200]
    class_2 = class_2[:200]
    class_1,class_2 = normalise(class_1,class_2)
    train = np.hstack((np.vstack((class_1.T,np.ones(200))),np.vstack((class_2.T,np.ones(200)))))
    X_train = train.T
    Y_train = np.hstack((labels_1,labels_2))
    X_test = np.hstack((np.vstack((test_1.T,np.ones(100))),np.vstack((test_2.T,np.ones(100))))).T
    Y_test = np.hstack((ytest_1,ytest_2))
#     print(labels_1.shape)
    w = np.array((np.random.normal(),np.random.normal(),1))
    alpha = 1e-3
    error_list = []
    test_list = []
    w_list = []
    Max_iters = 1000
    for i in range(Max_iters):
        loss = loss_function(w,X_train,Y_train)
        gd = calculate_gradient(w,X_train,Y_train)
        w_update = w-alpha*gd
        error_list.append(loss)
        w_list.append(w)
        test_preds = np.dot(X_test,w)
        test_list.append(200-np.count_nonzero(np.sign(test_preds) == Y_test))
        if loss < 1e-3:
            break
        w = w_update
    print("Gradient Descent After %d iterations"% i)
    print(loss)
    plt.plot(error_list)
    plt.show()
    plt.plot(test_list)
    plt.show()
        
main()
```

```python
###Q-5
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
import sys
from io import StringIO
```

```python
data = np.matrix(pd.read_excel("Concrete_Data.xls"))
X = data[:,:-1]
Y = data[:,-1]

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
sizes = [25,50,75]
activations = ['tanh','relu']
for sz in sizes:
    for act in activations:
        old_stdout = sys.stdout
        sys.stdout = mystdout = StringIO()
        model = MLPRegressor(hidden_layer_sizes=(sz),verbose=True,warm_start=True,activation=act,max_iter=1000)
        model.fit(X_train,np.ravel(Y_train))
        sys.stdout = old_stdout
        loss_history = mystdout.getvalue()
        loss_list = []
        for line in loss_history.split('\n'):
            if(len(line.split("loss = ")) == 1):
                continue
            loss_list.append(float(line.split("loss = ")[-1]))
        plt.plot(np.arange(len(loss_list)), loss_list)
        plt.xlabel("Time in epochs")
        plt.ylabel("Loss")
        plt.title(act+" activation with backpropagation and hidden nodes = "+str(sz))
        plt.show()
```
