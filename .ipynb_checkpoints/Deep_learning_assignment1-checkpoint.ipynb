```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from numpy import linalg as LA
```

```python
def plot_contour(X,Y,Error,history):
  fig,ax = plt.subplots(1,1)
  cp = ax.contour(X,Y,Error,colors = 'black',linestyles = 'dashed',linewidths = 0.5)
  ax.clabel(cp,inline = 1,fontsize = 10)
  cp = ax.contourf(X,Y,Error)
  fig.colorbar(cp)
  for i in range(len(history)-1):
    ax.annotate('',xy = history[i+1],xytext=history[i],arrowprops={'arrowstyle': 'fancy','color':'r','lw':0.5},va='center',ha='center')
  plt.show()
```

**Gradient Descent Method**

```python
x=np.random.uniform(-4,4,(2,400))
coeff=np.matrix([[2],[3]])
y = x.T*coeff  + np.random.normal(0,0.001,(400,1))
print(x.shape,y.shape)
```

```python
def objective_function(w1,w2,x,y):
  error = 0
  error=(w1*x[0]+w2*x[1] -y )**2
  # print( y, 2*x[0] + 3*x[1] + esp)
  return error
```

```python
w1 = np.linspace(-20,20,40)
w2 = np.linspace(-20,20,40)
W1,W2 = np.meshgrid(w1,w2)
Error=0
for i in range(0,400):
  Error += objective_function(W1,W2,x[:,i],y[i,0])
Error/=400
print(Error.shape)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(W1,W2,Error,color = 'blue')
ax.set_xlabel('w1')
ax.set_ylabel('w2')
ax.set_zlabel('Error')
plt.show()

```

```python
def com_gradient_error(w,x,y):
  y_pred=np.reshape(x.T@w,(400,1))
  # print(y_pred.shape)
  grad = 2*x@(y_pred - y)
  error = np.mean(np.square(y-y_pred))
  return grad,error

errors = []
history = []
w = np.reshape(np.array((-20,-20)),(2,1))
hes = 2*x@x.T # optimal learning rate
e_val, e_vec = LA.eig(hes)
lr_opt= 1/max(e_val) 
error=1
# print(lr_opt)
iter=0
while error>= 0.001:
  iter+=1
  gradient,error = com_gradient_error(w,x,y)
  print(w.shape, gradient.shape)
  new_w = w - lr_opt*gradient
  print("Iteration:",iter,"Error:",error)
  history.append(w)
  errors.append(error)
  if error < 0.001:
    print("Completed Gradient Descent")
    history.append(new_w)
    errors.append(error)
    break
  # print(w,"\n",new_w)
  w = new_w
```

```python
hes_inv = np.linalg.inv(2*np.dot(x,x.T)) # optimal learning rate
lr=[0.45,0.75,1,1.5]
for i in lr:
  history = []
  errors = []
  print("learning rate",i,"*optimal")
  w = np.reshape(np.array((-20,-20)),(2,1))
  error=1
  iter=0
  while error>= 0.001:
    iter+=1
    gradient,error = com_gradient_error(w,x,y)
    new_w = w - i*lr_opt*gradient
    # print("Iteration:",iter,"Error:",error)
    history.append([w[0,0],w[1,0]])
    errors.append(error)
    w = new_w
  history.append([new_w[0,0],new_w[1,0]])
  errors.append(error)
  print("Completed Gradient Descent")
  print()
  plot_contour(W1,W2,Error,history)
  plt.plot(errors)
  plt.show()



```

```python
def objective_function(X,Y):
	J = (X)**2 + 100.*(Y-X**2)**2
	return J

def cal_grad_error(w):
	x,y = w[0],w[1]
	dx = 2*x - 400*x*(y-x**2)
	dy = 200*(y - x**2)
	grad = np.array((dx,dy))
	error = objective_function(x,y)
	return grad,error 


x1 = np.arange(-2,1,0.02)
y1 = np.arange(-2,3,0.02)
X,Y = np.meshgrid(x1,y1)
Error = objective_function(X,Y)
w = np.random.rand(2, 1) 
alpha = 0.0013
history = []
errors = []
iter = 0
error=1
while error > 0.001:
  iter+=1
  gradient,error = cal_grad_error(w)
  new_w = w - alpha*gradient
  if iter%100== 0:
    print("Iteration:",iter,"Error:",error)
    history.append([w[0,0],w[1,0]])
    errors.append(error)
  w = new_w

errors.append(error)
history.append([w[0,0],w[1,0]])
print("Completed Gradient Descent in ",iter)
# print(w,iter)
plot_contour(X,Y,Error,history)
```
